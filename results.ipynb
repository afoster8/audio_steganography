```python
import matplotlib.pyplot as plt
import numpy as np
```

# SVM Results

```python
eight = [0.65, 0.81, 0.82, 0.81]
twenty = [0.76, 0.89, 0.93, 0.95]
thirty = [0.77, 0.89, 0.94, 1.0]
fifty = [0.69, 0.86, 0.96, 1.0]
sixty = [0.73, 0.89, 0.95, 1.0]

labels = ['8', '20', '30', '50', '60']
mfccs = ['2', '40', '800', '4000']

matrix = np.vstack((eight, twenty, thirty, fifty, sixty))

x = np.arange(len(labels))  
width = 0.15 

fig, ax = plt.subplots()
for i in range(len(matrix[0])):
    ax.bar(x + i*width, matrix[:, i], width, label=f'{mfccs[i]} Chars')

ax.set_xlabel('Number of MFCCs')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy by Number of MFCCs')
ax.set_xticks(x + width * (len(matrix[0]) - 1) / 2)
ax.set_xticklabels(labels)
ax.legend(loc='lower right')

plt.show()
```

```python
labels = ['2', '40', '800', '4000']
mfccs = ['8', '20', '30', '50', '60']
matrix = matrix.T

x = np.arange(len(labels))  
width = 0.15  

fig, ax = plt.subplots()
for i in range(len(matrix[0])):
    ax.bar(x + i*width, matrix[:, i], width, label=f'{mfccs[i]} MFCCs')

ax.set_xlabel('Number of Characters')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy by Number of Characters')
ax.set_xticks(x + width * (len(matrix[0]) - 1) / 2)
ax.set_xticklabels(labels)
ax.legend(loc='lower right')

plt.show()
```
